## Задание 1. Сосуд

```python
list_user, list_bound = [int(i) for i in input().split()], []
max_height = 0
# Поиск максимаьлного числа
for x in list_user:
    if max_height < x:
        max_height = x
        
# Перебираем эл-ты от 1 до максимального в списке и добавляем в новый список границы
for current_height in list_user: 
    for i in range(1, (current_height + 1)):
        if i >= current_height and i <= len(list_user):
            list_bound.append(1)
        elif i < current_height:
            list_bound.append(0)
            
    print(list_bound, end ='\t')
```
